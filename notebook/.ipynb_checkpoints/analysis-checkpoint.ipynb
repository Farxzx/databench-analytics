```python
import pandas as pd

employees = pd.read_csv("../employees.csv")
sales = pd.read_csv("../sales.csv")

print("Employees:", len(employees))
print("Sales:", len(sales))
```

```python
employees["department"].value_counts()
```

```python
import matplotlib.pyplot as plt

employees["department"].value_counts().plot(kind="bar")
plt.title("Employees by Department")
plt.show()
```

```python
top_sales = sales.groupby("employee_id")["sale_amount"].sum().sort_values(ascending=False).head(10)

top_sales.plot(kind="bar")
plt.title("Top 10 Employees by Total Sales")
plt.show()
```
